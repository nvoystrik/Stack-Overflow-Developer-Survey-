```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'C:\Users\nikvo\OneDrive\Documents\survey_results_public.csv')
df_schema=pd.read_csv(r'C:\Users\nikvo\OneDrive\Documents\survey_results_schema.csv')
```

```python
df=df.drop_duplicates()
```

```python
#Завдання 1. Підрахунок загальної кількості респондентів
total_respondents = len(df)
total_respondents
```

```python
#Завдання 2. Аналіз повноти відповідей респондентів
schema_questions = set(df_schema['qname']) #Список питань  у файлі survey_results_schema
survey_questions = set(df.columns) #Список   з основного датасету
questions_to_check = list(schema_questions.intersection(df.columns)) #ті запитання,які є в основному датасеті
complete_respondents = df.dropna(subset=questions_to_check) #видаляємо рядки з пропусками
count_resp=complete_respondents.shape[0] #
count_resp
```

```python
#Завдання 3. Статистичний аналіз досвіду респондентів
mean=df['WorkExp'].mean()
median=df['WorkExp'].median()
mode=df['WorkExp'].mode().iloc[0]
tabel=pd.DataFrame({
  'Міра': ['Mean', 'Median', 'Mode'],
  'Значення': [mean, median, mode]
})
tabel
```

```python
#Завдання 4. Аналіз віддаленої роботи
remote_count = df[df['RemoteWork'] == 'Remote'].shape[0]
remote_count
```

```python
#Завдання 5. Визначення популярності Python
#print(df['LanguageHaveWorkedWith'].unique())
python_mask = df['LanguageHaveWorkedWith'].str.contains('Python', na=False)
num_python = python_mask.sum()
percent_python = (num_python / len(df)) * 100
ppr=(num_python/df['LanguageHaveWorkedWith'].count())*100
print(f"Відсоток Python-програмістів від загальної к-ті респондентів : {percent_python:.2f}%")
print(f"Відсоток Python-програмістів серед тих,хто відповів : {ppr:.2f}%")

```

```python
#Завдання 6. Аналіз шляхів навчання програмуванню
num_online = df['LearnCode'].str.contains('Online Courses', na=False).sum()
total_learners = df['LearnCode'].count()

print(f"Абсолютна кількість учнів онлайн-курсів: {num_online}")
print(f"Відсоток онлайн-навчання серед респондентів: {(num_online/total_learners)*100:.2f}%")
```

```python
#Завдання 7. Географічний аналіз компенсації Python-розробників
python_users_df = df[df['LanguageHaveWorkedWith'].str.contains('Python', na=False)].copy()
# видаляємо тих, у кого не вказана зарплата
salary_df = python_users_df.dropna(subset=['ConvertedCompYearly', 'Country'])
country_salary_stats = salary_df.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median']).sort_values(by='median', ascending=False)
country_salary_stats.columns = ['Середня ($)', 'Медіана ($)']
print("Топ-10 країн за медіанною зарплатою Python-розробників:")
print(country_salary_stats.head(10))

```
